##                       Classification for "Nice Movie"

#### 简介：
    在一部电影上映以前，如何预测它是否会获得成功呢？或者说，电影投资者该如何筛选出可能获得非常好反响的剧本？电影制作公司该如何选择导演和演员？这些都需要十分复杂的数据分析，而我只进行一些简单分析，尝试从电影的历史数据中弄清是什么成就了一部好的电影。是题材？还是导演或者演员？还是投资多少？
    本notebook用机器学习的方式学习5000部电影的特征和评分，采用多种模型尝试得出分类方法并进行了比较。

#### 数据集：
    TMDB 5000 Movie Dataset from Kaggle

```python
def load_tmdb_movies(path):
    df = pd.read_csv(path)
    df['release_date'] = pd.to_datetime(df['release_date']).apply(lambda x:x.date())
    json_columns = ['genres', 'keywords', 'production_countries', 'production_companies', 'spoken_languages']
    
    for column in json_columns:
        df[column] = df[column].apply(json.loads)
        
    return df
```

```python
def load_tmdb_credits(path):
    df = pd.read_csv(path)
    json_columns = ['cast', 'crew']
    for column in json_columns:
        df[column] = df[column].apply(json.loads)
    return df
```

```python
def safe_access(container, indexs):
    # return NaN for missing value rather than error 
    result = container
    try:
        for i in indexs:
            result = result[i]
            
        return result
    except IndexError or KeyError:
        return np.nan
```

```python
def get_director(crew_data):
    directors = [x['name'] for x in crew_data if x['job'] == 'Director']
    return safe_access(directors, [0])
```

```python
def pipe_flattern_names(keywords):
    # turn a keywords list into a string split by "|"
    return '|'.join([x['name'] for x in keywords])
```

```python
def join_df(movies, credits):
    data = movies.copy()
    # join the two dataframe
    data['release_year'] = pd.to_datetime(movies['release_date']).apply(lambda x: x.year)
    # get first country name from json string
    data['country'] = movies['production_countries'].apply(lambda x : safe_access(x, [0, 'name']))
    # get language from json string
    data['language'] = movies['spoken_languages'].apply(lambda x : safe_access(x, [0, 'name']))
    # direcotr
    data['director'] = credits['crew'].apply(get_director)
    # main actors
    data["actor_1"] = credits['cast'].apply(lambda x: safe_access(x, [0, 'name']))
    data["actor_2"] = credits['cast'].apply(lambda x: safe_access(x, [1, 'name']))
    data["actor_3"] = credits['cast'].apply(lambda x: safe_access(x, [2, 'name']))
    data["actor_4"] = credits['cast'].apply(lambda x: safe_access(x, [3, 'name']))
    data["actor_5"] = credits['cast'].apply(lambda x: safe_access(x, [4, 'name']))
    #data = data[pd.notnull(data['actor_1'])]
    data['actor'] = data.actor_1
    data['actor'] = data['actor'] + data.actor_2.map(lambda s: ("|"+s) if pd.notnull(s) else "")
    data['actor'] = data['actor'] + data.actor_3.map(lambda s: ("|"+s) if pd.notnull(s) else "")
    data['actor'] = data['actor'] + data.actor_4.map(lambda s: ("|"+s) if pd.notnull(s) else "")
    data['actor'] = data['actor'] + data.actor_5.map(lambda s: ("|"+s) if pd.notnull(s) else "")
    data['actor'] = data['actor'].map(lambda s: s if pd.notnull(s) else "") # deal with NaN
    data['actor'] = data['actor'].apply(lambda s: s.encode("utf-8"))
    
    # company
    data["companies_1"] = movies["production_companies"].apply(lambda x: safe_access(x, [0, "name"]))
    data["companies_2"] = movies["production_companies"].apply(lambda x: safe_access(x, [1, "name"]))
    data["companies_3"] = movies["production_companies"].apply(lambda x: safe_access(x, [2, "name"]))
    data['companies'] = data.companies_1
    data['companies'] = data['companies'] + data.companies_2.map(lambda s: ("|"+s) if pd.notnull(s) else "")
    data['companies'] = data['companies'] + data.companies_3.map(lambda s: ("|"+s) if pd.notnull(s) else "")
   
    data['companies'] = data['companies'].map(lambda s: s if pd.notnull(s) else "")
    data['companies'] = data['companies'].apply(lambda s: s.encode("utf-8"))
    # genres
    data["genres"] = movies['genres'].apply(pipe_flattern_names)
    # plot_keywords
    data["keywords"] = movies["keywords"].apply(pipe_flattern_names)
    
    return data
                        
```

#### 我们看到keywords这一列就不太好对付了，这一列的关键词太多了，我们可以统计出所有关键词看一下。我们只选取出现频率的前20名构造特征向量。

```python
keyword_set = set()
for keyword_list in data['keywords'].str.split("|").values:
    if isinstance(keyword_list, float):  # keyword_list is NaN
        continue
    keyword_set = keyword_set.union(keyword_list)
keyword_set.remove("")   # remove ""
```

```python
def count_word(df, col, keyword_set):
    keyword_dict = dict()
    for keyword in keyword_set:
        keyword_dict[keyword] = 0
    for keyword_list in df[col].str.split("|").values:
        if isinstance(keyword_list, float):  # keyword_list is NaN
            continue
        for keyword in [x for x in keyword_list if x in keyword_set]:
            keyword_dict[keyword] += 1
            
    keyword_freq = []
    for key, freq in keyword_dict.items():
        keyword_freq.append([key, freq])
    
    keyword_freq.sort(key=lambda x:x[1], reverse = True)
    
    return keyword_freq, keyword_dict
```

```python
keyword_freq, keyword_dict = count_word(data, "keywords", keyword_set)
keyword_freq[:20]
keyword_list = pd.Series(keyword_dict)   # generate keyword dataframe]
keyword_list = keyword_list.sort_values(ascending=False)
keyword_list.head(40).plot(kind = "bar", legend = False, color = "blue", figsize = [15,8])
```

### 开始建模
#### 首先划分出训练集和检测集，比例7:3

```python
# prepare data set
X_trainset, X_testset, y_trainset, y_testset = train_test_split(X, Y, test_size=0.3, random_state=3)
print X_trainset.shape
print y_trainset.shape
print X_testset.shape
print y_testset.shape
```

#### 第一个模型，决策树

```python
from sklearn import metrics
tree = DecisionTreeClassifier(criterion="entropy")
tree.fit(X_trainset,y_trainset)
predTree = tree.predict(X_testset)
DT_acc = metrics.accuracy_score(y_testset, predTree)
print("DecisionTrees's Accuracy: "), DT_acc
```

#### 第二个模型，KNN

```python
knn = KNeighborsClassifier(n_neighbors= 60)
knn.fit(X_trainset, y_trainset)

training_accuracy = knn.score(X_trainset, y_trainset)
test_accuracy = knn.score(X_testset, y_testset)
print "Training Accuracy:", training_accuracy
print "Test Accuracy:", test_accuracy
KN_acc = test_accuracy
```

#### 第三个，梯度提升树

```python
gb = GradientBoostingClassifier(random_state=0, max_depth=1)
gb.fit(X_trainset, y_trainset)
GB_acc = gb.score(X_testset, y_testset)
print GB_acc
```

#### 第四个，随机森林

```python
Rfclf = RandomForestClassifier(random_state=0, max_depth=10, n_jobs = 5)
Rfclf.fit(X_trainset, y_trainset)
RF_acc = Rfclf.score(X_testset, y_testset)
print RF_acc
```

#### 第五个，朴素贝叶斯

```python
gnb = GaussianNB()
gnb.fit(X_trainset, y_trainset)
GNB_acc = gnb.score(X_testset, y_testset)
print GNB_acc
```

#### 我们来比较一下五中模型的准确率

```python
test_accuracy = []
test_accuracy.append(DT_acc)
test_accuracy.append(KN_acc)
test_accuracy.append(GB_acc)
test_accuracy.append(RF_acc)
test_accuracy.append(GNB_acc)

result = pd.DataFrame({"ACC":test_accuracy, "Model":["DecisionTree", "K-Newarest Neighbors"
                                                     , "Gradient Boosting", "Random Forest", "Naive Bayes"]})
graph = sns.barplot("ACC", "Model", data=result, palette="Set2", orient="h")
graph.set_xlabel("Accuracy")
graph = graph.set_title("Test Scores")
```

#### 从图中我们可以看出，准确率最高的两种分别是随机森林和迭代上升树，我们看一看在这两种模型中权重最高的几个特征列都是什么

```python
gb_most_important = set(gb_feature.sort_values(by=0, ascending=False).head(10).name.values)
rf_most_important = set(rf_feature.sort_values(by=0, ascending=False).head(10).name.values)
final = gb_most_important & rf_most_important
final
```
