## 数据收集练习
#### 运用本节课所学习到的内容，在网络上利用网络爬虫来收集数据。
要求：

  2. 如果你使用的是现成的数据集，那么请你编写利用爬虫来获取下面数据集的程序：   
     爬取下面这三个链接的网页中，列出的每一个Site的【名称、描述、网页地址】
     http://dmoztools.net/Computers/Programming/Languages/Python/Development_Tools/
     http://dmoztools.net/Computers/Programming/Languages/Python/Books/
     http://dmoztools.net/Computers/Programming/Languages/Python/Implementations/
  3. 无论你编写的程序爬取的是什么数据集，都应该把它们存储为.csv格式的文件，在后续课程的作业中可以复用这个数据集
  
  请在下面的代码框中编写你的程序。

```python
import scrapy
import csv
import os
from scrapy.crawler import CrawlerProcess


class MySpider(scrapy.Spider):
    name = "spider"

    def __init__(self):
        self.file = open("wangtao.csv", "w")
        self.csvWrite = csv.DictWriter(self.file,
                                       fieldnames=["name", "desc", "link"])
        item = {"name": "name", "desc": "desc", "link": "link"}
        self.csvWrite.writerow(item)

        # urls to crawl
        self.urls = []
        sites = ["Development_Tools", "Books", "Implementations"]
        for site in sites:
            self.urls.append("http://dmoztools.net/Computers/Programming/Languages/Python/" + site)

        print("The urls we are going to crawl are:\n")
        print(self.urls)

    def start_requests(self):
        for url in self.urls:
            yield scrapy.Request(url=url, callback=self.parse)

    def parse(self, response):
        title_and_desc_s = response.css("div.title-and-desc")
        for tad in title_and_desc_s:
            name = tad.css("div.site-title::text").extract_first()
            name = name.replace('\t', '').replace('\n', '').replace(' ', '')
            desc = tad.css("div.site-descr::text").extract_first()
            desc = desc.replace('\t', '').replace('\n', '').replace(' ', '').replace('\r', '')
            link = tad.css("a::attr(href)").extract_first()

            item = {"name": name, "desc": desc, "link": link}
            self.csvWrite.writerow(item)

        self.file.flush()
        os.fsync(self.file)
        print ("over: ")
        print (response.url)


process = CrawlerProcess({
    'USER_AGENT': 'Mozilla/4.0 (compatible; MSIE 7.0; Windows NT 5.1)'
})

process.crawl(MySpider)
process.start()
```
